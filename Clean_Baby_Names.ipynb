```python
def clean_ethnicity(input_ethnicity, lookfor_ethnicity: str, output_ethnicity: str):
    
    if input_ethnicity == lookfor_ethnicity:
        input_ethnicity = output_ethnicity
    
    return input_ethnicity
```

```python
df["Ethnicity"] = df["Ethnicity"].apply(lambda x: clean_ethnicity(x,"WHITE NON HISP", "WHITE NON HISPANIC"))
df.head()
```

```python
def clean_string(word: str):
    word = word.split(" ")
    complete_string = []
    
    for i in word:
        i = i.lower()
        first_letter = i[0].upper()
        complete_string.append(first_letter + i[1:])
        
    return " ".join(complete_string)
```

```python
df["Gender"] = df["Gender"].apply(clean_string)
df["Ethnicity"] = df["Ethnicity"].apply(clean_string)
df["Child's First Name"] = df["Child's First Name"].apply(clean_string)
```

```python
def filter_pre_year(dataframe_value, my_year: int):
    year_list = dataframe_value.tolist()
    counter = 0
    for year in year_list:
        if year == my_year:
            counter += 1
        else:
            continue
    return counter
```

```python
def filter_year(dataframe_value,chosen_year: int,prev_year: int):
    year_list = dataframe_value.tolist()
    counter = 0
    
    pre_year_count = filter_pre_year(dataframe_value,prev_year)
    
    for year in year_list:
        if year == chosen_year:
            counter += 1
        else:
            continue
    
    return pre_year_count, counter
```
